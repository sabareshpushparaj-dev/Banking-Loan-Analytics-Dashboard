```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
df = pd.read_csv(r"C:\Users\sabar\Downloads\Banking.csv")
df.head()
```

```python
## Generating Descriptive stats for the dataframe
df.describe()
```

```python
## Creating a Categorical column for Estimated income
bins = [0, 100000, 300000, float('inf')]
labels = ['Low', 'Med', 'High'] 
df['Income Band'] = pd.cut(df['Estimated Income'], bins=bins, labels=labels, right=False)
```

```python
df['Income Band'].value_counts().plot(kind='bar')
```

```python
## Examining the distribution of unique categories in categorical columns
categorical_cols = df[["BRId","GenderId", "IAId","Amount of Credit Cards", "Nationality", "Occupation", "Fee Structure", "Loyalty Classification", "Properties Owned","Risk Weighting", "Income Band"]]

for col in categorical_cols:
    
    display(df[col].value_counts())
    
```

```python
## Univariate Analysis
for i, predictor in enumerate(df[["BRId","GenderId", "IAId","Amount of Credit Cards", "Nationality", "Occupation", "Fee Structure", "Loyalty Classification", "Properties Owned","Risk Weighting", "Income Band"]]):
    plt.figure(i)
    sns.countplot(data=df,x=predictor)
```

```python
## Bivariate Analysis Based on Gender
for i, predictor in enumerate(df[["BRId","GenderId", "IAId","Amount of Credit Cards", "Nationality", "Occupation", "Fee Structure", "Loyalty Classification", "Properties Owned","Risk Weighting", "Income Band"]]):
    plt.figure(i)
    sns.countplot(data=df,x=predictor, hue='GenderId')
```

```python
## Numerical Analysis and Visulization

numerical_cols = ['Estimated Income','Superannuation Savings','Credit Card Balance','Bank Loans','Bank Deposits','Checking Accounts','Saving Accounts', 'Foreign Currency Account', 'Business Lending']
plt.figure(figsize=(20,8))
for i,col in enumerate(numerical_cols):
    plt.subplot(4,3,i+1)
    sns.histplot(df[col],kde=True)
    plt.title(col)
plt.show()
```

```python
## Heatmap

Correlation_Matrix = df[numerical_cols].corr()
plt.figure(figsize=(12,12))
sns.heatmap(Correlation_Matrix, annot=True, cmap='crest', fmt='.2f')
plt.title("Correlation Matrix")
plt.show()
```

Insights for the above EDA

The Strongest positive correlation occur among "Bank Deposits" with "Checking Account", "Saving Accounts" and "Foreign Currency Accounts"
Indicating that customers who maintain high balances in one account type often hold substantial amount/funds across other accounts as well.
